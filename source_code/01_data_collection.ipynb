```python
from google.colab import drive # connect to google drive
drive.mount("/content/drive")

project_path = "/content/drive/MyDrive/bee2041_empirical_project" # define project path
```

```python
from urllib.request import Request, urlopen # tools for sending a request to a website and retrieveing it's HTML content
import pandas as pd # library for data manipulation and tables
from io import StringIO # allows HTML strings to be treated as file objects for pandas so code still works in future versions
```

```python
url = "https://en.wikipedia.org/wiki/Taylor_Swift_albums_discography" # store the URL of wikipedia page

# create a request object with a browser-like header so that the request doesn't get blocked
request = Request(url, headers={"User-Agent": "Mozilla/5.0"})

page = urlopen(request) # open the web page
html_bytes = page.read() # read the HTML as bytes
html = html_bytes.decode("utf-8") # decode the bytes into a string

tables = pd.read_html(StringIO(html)) # extract the tables from the webpage into data frames
```

```python
wiki_studio_albums = tables[1].copy() # create a copy of the second table on the website

wiki_studio_albums.to_csv(f"{project_path}/data/raw/wiki_studio_albums.csv", index = False) # save raw data to drive

wiki_studio_albums.head() # show the first five rows
```

```python
wiki_tv_albums = tables[2].copy() # create a copy of the third table on the website

wiki_tv_albums.to_csv(f"{project_path}/data/raw/wiki_tv_albums.csv", index = False) # save raw data to drive

wiki_tv_albums.head() # show the first five rows
```
