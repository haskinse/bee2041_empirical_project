```python
from bs4 import BeautifulSoup  # parse HTML structure
from urllib.request import Request, urlopen # tools for sending a request to a website and retrieveing it's HTML content
import pandas as pd # library for data manipulation and tables
from io import StringIO # allows HTML strings to be treated as file objects for pandas so code still works in future versions
```

```python
url = "https://en.wikipedia.org/wiki/Taylor_Swift_albums_discography"  # Wikipedia discography page

request = Request(url, headers = {"User-Agent": "Mozilla/5.0"}) # send request with browser-like header to avoid being blocked
page = urlopen(request)

html = page.read().decode("utf-8") # read and decode HTML

soup = BeautifulSoup(html, "html.parser") # parse the page using BeautifulSoup to target specific elements

wiki_tables = soup.find_all("table", class_ = "wikitable") # find all tables on the page that use Wikipedia's 'wikitable' format

for i, table in enumerate(wiki_tables): # look at table captions to identify which ones contain album data
  caption = table.find("caption")
  print(i, caption.get_text(strip = True) if caption else "No caption")

# pick relevant tables
studio_table_html = str(wiki_tables[0])
tv_table_html = str(wiki_tables[1])

# convert tables into pandas dataframes
wiki_studio_albums = pd.read_html(StringIO(studio_table_html))[0]
wiki_tv_albums = pd.read_html(StringIO(tv_table_html))[0]
```

```python
def clean_album_name(s): # make album names easier to match
  return str(s).strip().lower() # return a lowercase string without extra whitespace
```

```python
studio_table_soup = BeautifulSoup(studio_table_html, "html.parser") # turn the HTML back into a beautiful soup object
tv_table_soup = BeautifulSoup(tv_table_html, "html.parser") # do it again for the taylor's version table

studio_titles = wiki_studio_albums[("Title", "Title")].astype(str).tolist() # extract the album names from the data frame
tv_titles = wiki_tv_albums[("Title", "Title")].astype(str).tolist() # "title" has to be repeated because the table has multiple heading layers

valid_album_names = {clean_album_name(name) for name in studio_titles + tv_titles} # combine both lists and slightly clean them for matching
```

```python
def get_album_links(table_soup, valid_album_names): # go through the table and get album names and links

  rows = [] # an empty list
  base_url = "https://en.wikipedia.org" # used in all links

  for row in table_soup.find_all("tr"): # loop through each row in the table

    title_cell = row.find("th") # album titles are normally in the header cell

    if not title_cell: # if there is no title cell
      continue # skip row

    link = title_cell.find("a", href = True) # look for a link in that cell

    if not link: # if there is no linke
      continue # skip

    if not link["href"].startswith("/wiki/"): # check it's a proper wikipedia page
      continue # skip if it's not

    album_name = link.get_text(strip = True) # get the visible album name

    if clean_album_name(album_name) in valid_album_names: # check that it's a valid album name
      rows.append({"album_name": album_name,"album_url": base_url + link["href"]}) # add it to the list

  return pd.DataFrame(rows) # turn the list into a dataframe
```

```python
studio_links = get_album_links(studio_table_soup, valid_album_names) # get the links for the studio albums
tv_links = get_album_links(tv_table_soup, valid_album_names) # get the links for the taylor's version albums

album_links = pd.concat([studio_links, tv_links], ignore_index = True) # combine the two to get links for all albums

album_links.head() # show first five rows
```

```python
def scrape_album_page_info(url): # visit each album page and get genre and page length

  request = Request(url, headers = {"User-Agent": "Mozilla/5.0"}) # just like earlier
  page = urlopen(request) # go to website

  html = page.read().decode("utf-8") # read website content
  soup = BeautifulSoup(html, "html.parser") # structures HTML

  page_text = soup.get_text(" ", strip = True) # get everything written on a page

  page_length = len(page_text) # count how long the page text is

  infobox = soup.find("table", class_= "infobox") # look for the page's infobox

  if infobox: # if the infobox exists

    for row in infobox.find_all("tr"): # go through each row in the infobox

      header = row.find("th") # left side is the label
      value = row.find("td") # right side is the value

      if not header or not value: # skip rows that aren't proper labels with values
        continue

      header_text = header.get_text(" ", strip = True) # get the text from the header

      if header_text == "Genre": # check it's the genre

        genre = value.get_text(", ", strip=True) # get the genre text
        break # stop once it has been found

    return genre, page_length # return information
```

```python
import time # used to pause between requests
```

```python
page_rows = [] # an empty list to store information found

for i, row in album_links.iterrows(): # for each fow in the data frame

    genre, page_length = scrape_album_page_info(row["album_url"]) # scrape the page for genre and length

    page_rows.append({"album_name": row["album_name"], "genre": genre, "page_length": page_length}) # add information to the list

    time.sleep(0.5) # add a small delay to avoid spamming
```

```python
album_page_info = pd.DataFrame(page_rows) # convert list to a data frame

album_page_info.to_csv(f"{project_path}/data/raw/wiki_album_page_info.csv", index = False)

album_page_info.head() # show the first five rows
```

```python
client_id = os.getenv("SPOTIFY_CLIENT_ID") # store client id for authentication
client_secret = os.getenv("SPOTIFY_CLIENT_SECRET") # store client secret for authentication

client_credentials_manager = SpotifyClientCredentials(client_id = client_id, client_secret = client_secret) # create an authentication manager

sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager) # create an API client for making requests
```

```python
import time # used to delay code (handle API rate limits)
```

```python
# search for taylor swift and pull out the first line
artist = sp.search(q = "artist:Taylor Swift", type = "artist", limit = 1)["artists"]["items"][0]

artist_id = artist["id"] # stores spotify artist id for API later
artist_name = artist["name"] # stores spotify artist name

all_albums = [] # make an empty list to store all album data
offset = 0 # set offset for paigination (spotify API can only return 10 results)

while True: # until stopped
  batch = sp.artist_albums(artist_id, album_type = "album", limit = 10, offset = offset)["items"] # retrieve a batch of albums

  if not batch: # if there are no more items in the batch
    break # break the loop

  else: # otherwise...
    all_albums.extend(batch) # add the new batch of albums to the albums list
    offset += 10 # move forward by ten albums (to get the next batch)
    time.sleep(0.5) # wait 0.5 seconds so the API isn't overwhelmed

rows = [] # make an empty list to store rows for the track data frame

for album in all_albums: # for each album in the list

  time.sleep(0.3)  # wait 0.3 seconds to avoid the rate limit

  tracks = sp.album_tracks(album["id"]) # retrieve track information for the current album

  for track in tracks["items"]: # for each track
    artist_list = [a["name"] for a in track["artists"]] # take the artists' names and store them in a list

    # add a dictionary to represent the row's data
    rows.append({"artist_name": ", ".join(artist_list), "album_name": album["name"], "release_date": album["release_date"], "track_name": track["name"], "track_number": track["track_number"], "explicit": track["explicit"]})

spotify_data = pd.DataFrame(rows) # read the list of rows into a data frame

spotify_data.to_csv(f"{project_path}/data/raw/spotify_data.csv", index = False) # save raw data to drive index = False) # store as a csv to avoid having to repeatedly use the spotify API

spotify_data.head() # show first five rows
```

```python
all_artists = [] # create an empty list to store artists

for artists in spotify_data["artist_name"]: # for each row's artist name column
  split_artists = artists.split(",") # split the artist names where there is a comma
  split_artists = [a.strip() for a in split_artists] # strip the whitespace
  all_artists.extend(split_artists) # add each split artist to the list

unique_artists = sorted(set(all_artists)) # sort the list and make it a set
```

```python
import requests # used to send HTTP requests to APIs like Last.fm
```

```python
def get_lastfm_stats(artist_name, api_key, base_url): # retrieve listener and playcount statistics for a given artist from the Last.fm API

  # define parameters for artist search
  search_params = {"method": "artist.search", "artist": artist_name, "api_key": api_key, "format": "json", "limit": 1}

  search_data = requests.get(base_url, params = search_params).json() # send request to Last.fm API and parse JSON response

  matches = search_data["results"]["artistmatches"]["artist"] # extract list of matching artists from search results

  matched_name = matches[0]["name"]  # pick the top match to ensure consistent artist identification

  # define parameters to retrieve detailed artist information
  info_params = {"method": "artist.getInfo", "artist": matched_name, "api_key": api_key, "format": "json", "autocorrect": 0}

  info_data = requests.get(base_url, params = info_params).json() # request artist information from Last.fm

  stats = info_data["artist"]["stats"] # extract listener and playcount statistics

  return stats["listeners"], stats["playcount"] # return metrics for analysis
```

```python
os.environ["LASTFM_API_KEY"] = "lastfm_api_key" # store client id for authentication

# credentials removed for security - see README
```

```python
api_key = os.getenv("LASTFM_API_KEY") # API key retrieved securely from environment variables
# credentials removed for security - see README

base_url = "https://ws.audioscrobbler.com/2.0/" # store the base url for Last.fm API

# manually specified stats for artists where API results are inconsistent or incorrect
manual_stats = {"Florence + The Machine": {"listeners": 3891116, "playcount": 280600000}}

rows = [] # create list to store rows of data for artist data frame

for artist_name in unique_artists: # for each artist

  if artist_name in manual_stats: # use manually defined values to override unreliable API results
    listeners = manual_stats[artist_name]["listeners"]
    playcount = manual_stats[artist_name]["playcount"]

  else: # retrieve listener and playcount data from Last.fm API
    listeners, playcount = get_lastfm_stats(artist_name, api_key, base_url)

  rows.append({"artist_name": artist_name, "listeners": listeners, "playcount": playcount}) # append artist statistics as a row in the dataset

  time.sleep(0.3) # pause between requests to avoid overwhelming API
```

```python
lastfm_data = pd.DataFrame(rows) # create a data frame our of the rows

lastfm_data.to_csv(f"{project_path}/data/raw/lastfm_data.csv", index = False) # save raw data to drive

lastfm_data.head() # show first five rows
```
